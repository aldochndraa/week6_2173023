```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
np.random.seed(2)
data=np.random.normal(loc=100,scale=10,size=200)
data
```

```python
plt.boxplot(data,labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

```python
plt.boxplot(data,labels=['Data'], notch=True)

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')


plt.show()
```

```python
plt.boxplot(data,labels=['Data'],showfliers=True,flierprops={'markerfacecolor':'r','marker':'d'})

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')


plt.show()
```

```python
plt.boxplot(data,labels=['Data'],vert=False)
plt.title('Horisontal Boc Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')


plt.show()
```

```python
np.random.seed(10)
data1=np.random.normal(100,10,200)
data2=np.random.normal(80,20,100)
data3=np.random.normal(90,20,200)
data4=np.random.normal(100,20)
data=[data1,data2,data3,data4]
labels =['Data 1', 'Data 2','Data 3','Data 4']
```

```python
plt.boxplot(data, labels=labels)

plt.title(' Multiple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.show()
```

```python
np.random.seed(2)
data=np.random.normal(loc=100,scale=10,size=200)
data
```

```python
plt.violinplot(data)
plt.title('Simple Violoin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.show()
```

```python
plt.violinplot(data,
               showextrema=True,
               showmeans=False,
               showmedians=False,
               quantiles=None)
plt.title('Violoin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.show()
```

```python
plt.violinplot(data, vert=False)

plt.title('Horizontal Violoin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.show()
```

```python
x=np.linspace(1,10,25)
x
```

```python
fig,ax1 = plt.subplots()

ax1.plot(x,np.exp(x),'bs-',label='exp')
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('exp')

ax2=ax1.twinx()
ax2.plot(x,np.log(x),'ro-',label='log')
ax2.set_ylabel('log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc='upper left')

plt.show()
```

```python
x=np.random.randint(16,40,size=30)
x
```

```python
def C2F(celcius=0):
    return (celcius+1.8)+32
def konversi_sumbu(ax1):
    y1,y2=ax1.get_ylim()
    ax2.set_ylim(C2F(y1), C2F(y2))
    ax2.figure.canvas.draw()

fig, ax1=plt.subplots()
ax2=ax1.twinx()

ax1.callbacks.connect('ylim_changed', konversi_sumbu)
ax1.plot(x)

ax1.set_xlabel('Hari')
ax1.set_ylabel('Celcius')
ax2.set_ylabel('Fahrenheit')

fig.suptitle('Tempratur Udara')
plt.show()
```
